# Seaborn Tutorial

```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

## Joint Plot

### Univariate analysis

```python
sns.jointplot(x='tip', y='total_bill',data=df,kind='reg')
```

### pair plot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

### Dist plot

```python
sns.distplot(df['tip'])
```

# Tutorial 10 - Seaborn- Part- 2

## Catagorical Plots

```python
## Count plot

sns.countplot('day',data=df)
```

```python
## Bar plot

sns.barplot(x='sex',y='total_bill',data=df)
```

```python
## Box plot

sns.boxplot('sex','total_bill', data=df)
```

```python
sns.boxplot(x='day', y='total_bill', data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot(x='total_bill', y='day', hue='smoker', data=df)
```

```python
## violin plot

sns.violinplot(x='total_bill', y='day',data=df)
```
