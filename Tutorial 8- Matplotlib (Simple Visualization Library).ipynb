```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
```

```python
## Simple Example

x=np.arange(0,10)
y=np.arange(11,21)
```

```python
## plotting using matlpotlib
## plt scatter

plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')
```

```python
## plt plot
y=x*x
plt.plot(x,y,"r*-")

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
```

```python
## Creating Subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*-')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
```

```python
x= np.arange(0,11)
y=3*x+5
plt.title("Matplotlib Demo")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(x,y)
plt.show()
```

```python
# Compute the x and y coordinates for points on a sine curve
x= np.arange(0,4 * np.pi, 0.1)
y= np.sin(x)
plt.title("sine wave form")

# plot the points using matplotlib
plt.plot(x,y)
plt.show()
```

```python
#3 Bar plot

x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]
plt.bar(x, y)
plt.bar(x2 , y2, color = 'g')
plt.title('Bar Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

## Histograms

```python
a= np.array([22,87,5,43,56,74,99,86,45,5,22,27,30,37])
plt.hist(a)
plt.title("histogram")
plt.show()
```

## Box Plot using Matplotlib

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data, vert=True,patch_artist=True);
```

## Pie Chart

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)

#plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()
```
